```julia
using MeshTools
using MeshViz, CairoMakie
```

```julia
outer = [(0.0,0.0),(1.0,0.0),(1.0,1.0),(0.0,1.0),(0.0,0.0)] # last point is optional
hole1 = [(0.2,0.2),(0.4,0.2),(0.4,0.4),(0.2,0.4),(0.2,0.2)]
hole2 = [(0.6,0.2),(0.8,0.2),(0.8,0.4),(0.6,0.4),(0.6,0.2)]

mesh = create_mesh(outer; inner = [hole1, hole2], outer_length=0.1, inner_lengths = (0.1, 0.1), spline=false, spline_inners=(false,false))
```

```julia
viz(mesh,showfacets=true)
```

```julia
mesh = create_mesh(outer; inner = [hole1, hole2], outer_length=0.1, inner_lengths = (0.01, 0.5), spline=true, spline_inners=(true,false))
viz(mesh,showfacets=true)
```
