```julia
using Meshes, MeshViz, CairoMakie, MeshTools
```

```julia
outer = [(0.0,0.0),(1.0*2pi,0.0),(1.0*2pi,1.0*2pi),(0.0,1.0*2pi),(0.0,0.0)] # last point is optional

mesh = create_mesh(outer; outer_length=0.15)

h(x,y) = sin(x)*cos(y) # or h(p::Point2) = ...
fvals = [h(x,y) for (x,y) in Meshes.coordinates.(vertices(mesh))]

fig, ax = viz(mesh,showfacets=true,color=fvals)

ctrs = MeshTools.contour(h, mesh, -0.1)
for c in ctrs
    x = getindex.(c,1)
    y = getindex.(c,2)
    lines!(ax,x,y,linewidth=5,color=:pink)
end
fig
```
