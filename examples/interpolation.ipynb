## Barycentric Interpolation on a Triangle

```julia
t = Triangle((0.0,0.0),(1.0,0.0),(0.5,1.0))

x = range(0,1,length=50)
y = range(0,1,length=50)
λ₁ = Float64[]
λ₂ = Float64[]
λ₃ = Float64[]
F = Float64[]
for xx in x, yy in y
    p = Point(xx,yy)
    if p ∈ t
        l1,l2,l3 = barycentric_coordinates(t,p)
        push!(λ₁, l1)
        push!(λ₂, l2)
        push!(λ₃, l3)
        f = interpolate(t,[1.0,2.0,3.0], p)
        push!(F,f)
    end
end

fig = ff.create_ternary_contour(hcat(λ₁, λ₂, λ₃)', F,ncontours=30, showscale=true,colorscale="Hot",pole_labels=["1","2","3"])
fig
```

## Interpolation on Mesh

```julia
outer = [(0.0,0.0),(1.0,0.0),(1.0,1.0),(0.0,1.0)]

mesh = create_mesh(outer; outer_length=0.02, spline=false)

f(x,y) = cos(5x)*sin(2y)

F = [f(coordinates(p)...) for p in vertices(mesh)];

points = [rand(2) for i=1:100]
Fp = [interpolate(mesh,F,p) for p in points]
Fa = [f(p...) for p in points]
err = sum(abs.(Fp .- Fa))/length(points)
println("Average Error: $(err)")
```
